Para resolver el item (c) de las preguntas en la imagen anexada, debemos demostrar que la ecuación (8) se puede escribir matricialmente como $ y = F^{-1} c $, donde $ y = (y_0, \ldots, y_{n-1})^T $, $ c = (c_0, \ldots, c_{n-1})^T $, y $ F^{-1} $ es la inversa de la matriz $ F $ definida previamente. Usaremos la información del enunciado y los resultados anteriores.
Razonamiento:

Definición de la ecuación (8):

Según el enunciado, la ecuación (8) de la Transformada Inversa Discreta de Fourier (IDFT) se da por:
$ y_k = f(x_k) = \sum_{j=0}^{n-1} c_j \omega^{-jk}, \quad k = 0, 1, \ldots, n-1, $
donde $ \omega = e^{-2\pi i / n} $ y $ \omega^{-jk} = (e^{-2\pi i / n})^{-jk} = e^{2\pi i jk / n} $.
Esto indica que cada componente $ y_k $ es una combinación lineal de los coeficientes $ c_j $, con coeficientes dados por $ \omega^{-jk} $.


Relación con la matriz $ F $:

Del item (a), la matriz $ F $ asociada a la Transformada Discreta de Fourier (DFT) es:

$$
F = \frac{1}{n} \begin{bmatrix}
1 & \omega & \omega^2 & \cdots & \omega^{n-1} \\
1 & \omega^2 & \omega^4 & \cdots & \omega^{2(n-1)} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & \omega^{n-1} & \omega^{2(n-1)} & \cdots & \omega^{(n-1)(n-1)}
\end{bmatrix},
$$
y el producto $$  c = F y  $$ se deriva de la ecuación (7): $$  c_j = \frac{1}{n} \sum_{k=0}^{n-1} y_k \omega^{jk}  $$.

Del item (b), se demostró que $ F^H F = \frac{1}{n} I $, lo que implica $ F^{-1} = n F^H $, donde $ F^H $ es la transpuesta conjugada de $ F $.


Construcción de $ F^{-1} $:

La matriz $ F^H $ tiene elementos $ (F^H)_{jk} = \bar{f}_{kj} = \frac{1}{n} \omega^{-jk} $ (como se derivó en el item (b)).
Por lo tanto, $ F^{-1} = n F^H $ tiene elementos:
$ (F^{-1})_{jk} = n \cdot \frac{1}{n} \omega^{-jk} = \omega^{-jk}. $
Así, la matriz $ F^{-1} $ es:
$$
F^{-1} = \begin{bmatrix}
1 & \omega^{-1} & \omega^{-2} & \cdots & \omega^{-(n-1)} \\
1 & \omega^{-2} & \omega^{-4} & \cdots & \omega^{-2(n-1)} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & \omega^{-(n-1)} & \omega^{-2(n-1)} & \cdots & \omega^{-(n-1)(n-1)}
\end{bmatrix},$$




Verificación de $ y = F^{-1} c $:

El producto $ y = F^{-1} c $ da como resultado $ y_k = \sum_{j=0}^{n-1} (F^{-1})_{kj} c_j $.
Sustituyendo $ (F^{-1})_{kj} = \omega^{-jk} $:
$ y_k = \sum_{j=0}^{n-1} \omega^{-jk} c_j, $
que coincide exactamente con la ecuación (8) $ y_k = \sum_{j=0}^{n-1} c_j \omega^{-jk} $.
Esto confirma que la transformación inversa se puede expresar matricialmente como $ y = F^{-1} c $.
