Trayendo un resultado bsico de lgebra lineal estudiado en clases:

Si $\mathcal{V}$ es un espacio con producto interno $\langle \cdot, \cdot \rangle$ y si $\mathcal{B} = \{\varphi_0, \varphi_1, \ldots, \varphi_{n-1}\}$ es una base ortonormal de $\mathcal{V}$; entonces cada $f \in \mathcal{V}$ se puede escribir como
\begin{equation}
    f = \langle \varphi_0, f \rangle \varphi_0 + \langle \varphi_1, f \rangle \varphi_1 + \cdots + \langle \varphi_{n-1}, f \rangle \varphi_{n-1}.
\end{equation}


Pero este interpoolador en la prctica no nos sirve de mucho, pues construirlo a partir del producto interno dado sera sumamente costoso tomando en cuenta adems a todos los nodos de la base $\mathcal{B}$. Tendramos que integrar sobre el dominio completo de la funcin y adems asumiendo que conocemos la funcin completa y que es si quiera integrable. Es un resultado terico perfecto pero que no nos dice nada sobre la construccin del interpolador a partir de datos discretos.
