Trayendo un resultado básico de álgebra lineal estudiado en clases:

Si $\mathcal{V}$ es un espacio con producto interno $\langle \cdot, \cdot \rangle$ y si $\mathcal{B} = \{\varphi_0, \varphi_1, \ldots, \varphi_{n-1}\}$ es una base ortonormal de $\mathcal{V}$; entonces cada $f \in \mathcal{V}$ se puede escribir como
\begin{equation}
    f = \langle \varphi_0, f \rangle \varphi_0 + \langle \varphi_1, f \rangle \varphi_1 + \cdots + \langle \varphi_{n-1}, f \rangle \varphi_{n-1}.
\end{equation}


Pero este interpoolador en la práctica no nos sirve de mucho, pues construirlo a partir del producto interno dado sería sumamente costoso tomando en cuenta además a todos los nodos de la base $\mathcal{B}$. Tendríamos que integrar sobre el dominio completo de la función y además asumiendo que conocemos la función completa y que es si quiera integrable. Es un resultado teórico perfecto pero que no nos dice nada sobre la construcción del interpolador a partir de datos discretos.
