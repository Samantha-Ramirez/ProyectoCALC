Recordemos que las funciones $\phi_j$ vienen dadas por $\phi_j := \phi_j(x)= e^{ijx}, j = 0, 1, \dots, n$

Puesto que $$ forma un conjunto ortonormal, no es necesario normalizar en el producto interno. Adems, llamaremos para evitar confusin con la notacin al ndice de la sumatoria como $\overline{k}$ para evitar confusin con el subndice de la funcin. Con lo anterior nos queda:

$$\langle \phi_j, \phi_k \rangle = \sum_{\overline{k}=0}^{n-1}\phi_j\overline{\phi_k} =
\sum_{\overline{k}=0}^{n-1}e^{ijx_k}*e^{-ikx_k}=\sum_{\overline{k}=0}^{n-1}e^{i(j-k)x_k}$$

Recordando la definicin de los $x_k$

$$\langle \phi_j, \phi_k \rangle = \sum_{\overline{k}=0}^{n-1}e^{i(j-k)\frac{2\pi \overline{k}}{n}}$$

Si $j - k$ es divisible entre $n$ entonces al simplificar obtenemos:

$$\langle \phi_j, \phi_k \rangle = \sum_{\overline{k}=0}^{n-1}e^{i2\pi \overline{k}}=
\sum_{\overline{k}=0}^{n-1}\mathcal{cos(2\pi \overline{k})} + i\mathcal{sin(2\pi \overline{k})}=$$

$$\sum_{\overline{k}=0}^{n-1}1 = n$$

Para cualquier otro caso, podemos escribir la sumatoria como una serie geomtrica y llamaremos $j - k$ como $m$:

$$\langle \phi_j, \phi_k \rangle = \sum_{\overline{k}=0}^{n-1}e^{im\frac{2\pi \overline{k}}{n}}  = \sum_{\overline{k}=0}^{n-1}\left( e^{im2\pi/n} \right)^{\overline{k}}=$$

$$S = \frac{1-\left( e^{im2\pi/n} \right)^{n}}{1-\left( e^{im2\pi/n}  \right)}=$$

$$\frac{1-\left( e^{n(im2\pi/n)} \right)}{1-\left( e^{im2\pi/n}  \right)}=
\frac{1-\left( e^{im2\pi)} \right)}{1-\left( e^{im2\pi/n}  \right)}$$

Siguiendo el argumento usado en el caso anterior, nos queda que $e^{im2\pi} = 1$, luego para el segundo caso podemos concluir que

$$\langle \phi_j, \phi_k \rangle = \frac{1-1}{1-e^{im2\pi/n}} = 0$$
