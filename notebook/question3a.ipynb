Probar Positividad:

$$\left\langle x,x\right\rangle = \frac{1}{n}∑_{k=0}^{n-1}\mathcal{f(x_k)}\overline{\mathcal{f(x_k)}}$$

Pero una función multiplicada por su conjugada es siempre mayor o igual a cero:
$$z\overline{z}=(a+bi)(a-bi)=a^{2}-(bi)^{2}=a^{2}-b^{2}i^{2}=a^{2}-b^{2}(-1)=a^{2}+b^{2}$$

Dado que el producto interno es una suma de número positivos dividida por un cociente positivo, es claro que el producto interno $\langle x,x \rangle \ge 0$

Sin embargo, las tres propiedades dadas, la segunda parte de la primera no se cumple, pues aunque es posible demostrar que para cualquier función igual a cero en todos los puntos $x_k$ utilizados, el producto interno es cero, el caso contrario no es cierto. A continuación el contraejemplo:

Tomando como caso particular a la función $f=\mathcal{sin}$, es posible seleccionar los nodos $x_k$ que anulen la función pues esta tiene puntos críticos en los múltiplos de $π$ a partir del cero. Para construir un contraejemplo tomamos $n = 2$ y puesto que los nodos vienen dados por $x_k = 2\pi k/n$, el conjunto resultante es $x_k = \{0, π\}$. Luego:

$$\langle f,f \rangle = \frac{1}{2}\left( \mathcal{sin(0)}\overline{\mathcal{sin(0)}} + \mathcal{sin(π)}\overline{\mathcal{sin(π)}} \right) = 0$$

Pero $\mathcal{f(π/2)}=1 ⇒ f \neq 0$

Probar conjugación:

$$\overline{\langle y,x \rangle} = \overline{\frac{1}{n}∑_{k=0}^{n-1}\mathcal{y}\overline{x}} = \frac{1}{n}∑_{k=0}^{n-1}\overline{y \overline{x}} = \frac{1}{n}∑_{k=0}^{n-1}x\overline{y}=\langle x,y \rangle$$

Probar Aditividad y Homogeneidad:

Recordando que el conjugado de la suma al igual a la suma de los conjugados y el conjugado de una constante real es la misma constante real:

$$\langle f, \alpha g + \beta h \rangle = \frac{1}{n}∑_{k=0}^{n-1}f\overline{\alpha g + \beta h}=$$

$$ \alpha\frac{1}{n}∑_{k=0}^{n-1}f\overline{g} + \beta\frac{1}{n}∑_{k=0}^{n-1}f\overline{h} = \alpha \langle f,g \rangle + \beta \langle f,h \rangle$$

Probando así la Aditividad y Homogeneidad del producto interno dado
