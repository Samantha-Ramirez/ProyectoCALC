Probar Positividad:

$$\left\langle x,x\right\rangle = \frac{1}{n}_{k=0}^{n-1}\mathcal{f(x_k)}\overline{\mathcal{f(x_k)}}$$

Pero una funcin multiplicada por su conjugada es siempre mayor o igual a cero:
$$z\overline{z}=(a+bi)(a-bi)=a^{2}-(bi)^{2}=a^{2}-b^{2}i^{2}=a^{2}-b^{2}(-1)=a^{2}+b^{2}$$

Dado que el producto interno es una suma de nmero positivos dividida por un cociente positivo, es claro que el producto interno $\langle x,x \rangle \ge 0$

Sin embargo, las tres propiedades dadas, la segunda parte de la primera no se cumple, pues aunque es posible demostrar que para cualquier funcin igual a cero en todos los puntos $x_k$ utilizados, el producto interno es cero, el caso contrario no es cierto. A continuacin el contraejemplo:

Tomando como caso particular a la funcin $f=\mathcal{sin}$, es posible seleccionar los nodos $x_k$ que anulen la funcin pues esta tiene puntos crticos en los mltiplos de $$ a partir del cero. $$\langle f,f \rangle = \frac{1}{2}\left( \mathcal{sin(0)}\overline{\mathcal{sin(0)}} + \mathcal{sin()}\overline{\mathcal{sin()}} \right) = 0$$
