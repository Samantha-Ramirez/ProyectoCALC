1. Siendo $V$ el espacio de las funciones periódicas con $p = 2π$, definidas en el espacio completo
2. En este espacio, el conjunto $\mathcal{B} \subset \mathcal{V} \text{ definido como } \mathcal{B} = \{\phi_0, \phi_1, \ldots, \phi_{n-1}\}$ forma una base ortonormal donde:
  \begin{equation}
  \phi_j := \phi_j(x) = e^{ijx}, \text{ para } j = 0, 1, \ldots, n.
  \end{equation}
siendo $i$ la unidad compleja, es decir, $i = \sqrt{-1}$; mientras que el producto interno usado se define como (3)
  \begin{equation}
  \langle f, g \rangle = \frac{1}{2\pi} \int_{-\pi}^{\pi} f(t) \overline{g(t)} dt,
  \end{equation}
donde $\overline{g(t)}$ denota al transpuesto conjugado de $g(t)$.



B es ortonormal si y sólo si para cada par $f, g ∈ B$ se cumple que:
$$f \neq g ⇒ <f,g> = 0$$
$$f = g ⇒ <f,g> = 1$$

Estudiemos el caso con las funciones $\phi_j, \phi_k$ para el producto interno dado:

$$<\phi_j, \phi_k> = \frac{1}{2π}∫_{-π}^{π}{\phi_j,\overline{\phi_k}}\,dx=$$
$$\frac{1}{2π}∫_{-π}^{π}{e^{ijx}*e^{-ikx}}\,dx=\frac{1}{2π}∫_{-π}^{π}{e^{i(j-k)x}}\,dx=$$


Para $j \neq k$, decimos que $j-k$ es una constante cualquiera $m$:

$$<\phi_j, \phi_k> = \frac{1}{2π}∫_{-π}^{π}{e^{imx}}\,dx=$$

$$\frac{1}{2π}\left.\left[ \frac{e^{imx}}{imx}\right]\right|_{-π}^{π}=$$

$$\frac{1}{2π}\left( \frac{e^{imxπ}-e^{-imxπ}}{imx} \right)$$

Y sabemos por Euler que $e^{i\theta} - e^{-i\theta} = 2\sin{(θ)}$, de donde se desprende que:

$$<\phi_j, \phi_k> = \frac{1}{2π}\frac{2i\sin{(mπ)}}{imx}$$

Y ya que la función seno se anula en todos los múltiplos enteros de π, comenzando desde 0, este último resultado es igual a 0 probando así el primer caso.

Para $j=k$, su diferencia es 0 y nos queda:

$$<\phi_j, \phi_k> = \frac{1}{2π}∫_{-π}^{π}{e^{0}}\,dx=$$

$$<\phi_j, \phi_k> = \frac{1}{2π}∫_{-π}^{π}\,dx=$$

$$\frac{1}{2π}2π = 1$$
