

Si escribimos los coeficientes como $c = (c_0, \ldots,c_{j},\ldots, c_{n-1})^T$ y $y = (y_0, \ldots,y_{j},\ldots, y_{n-1})^T$
$$
(7) \quad c_j = \frac{1}{n} \sum_{k=0}^{n-1} y_k \omega^{jk}
\\c_j=\frac{1}{n}(y_{0}w^{j0} + \dots + y_{j}w^{jj} + \dots + y_{n-1}w^{j(n-1)})
\\c_j=
y_{0}f_{j0} + \dots + y_{j}f_{jj} + \dots + y_{n-1}f_{j(n-1)}
$$
Queremos escribir la Transformada de Fourier (7) como un producto matriz vector.
$$
c=
\begin{bmatrix}
y_{0}f_{00} + \dots + y_{n-1}f_{0(n-1)}\\
\vdots \\
y_{0}f_{j0} + \dots + y_{n-1}f_{j(n-1)}\\
\vdots \\
y_{0}f_{(n-1)0} + \dots + y_{n-1}f_{(n-1)(n-1)}
\end{bmatrix}

= y_0 \begin{bmatrix} 
f_{00} \\ \vdots \\ f_{j0} \\\vdots \\ f_{(n-1)0}
\end{bmatrix} + \dots +  y_{n-1}\begin{bmatrix} 
f_{0(n-1)} \\\vdots \\ f_{j(n-1)} \\\vdots \\ f_{(n-1)(n-1)}
\end{bmatrix}
\\c=
\begin{bmatrix} 
f_{00} & \dots & f_{0(n-1)} \\ \vdots & \vdots & \vdots \\ f_{j0} & \dots & f_{j(n-1)}\\\vdots & \vdots & \vdots  \\ f_{(n-1)0} & \dots & f_{(n-1)(n-1)}
\end{bmatrix}
y
\implies 
c=Fy
$$
Donde la matriz $F$ (llamada Matriz de Fourier) tiene dimensiones $n \times n$ y cada elemento $ f_{jk} $ es $ \frac{1}{n} \omega^{jk} $. Finalmente,
$$  F = \frac{1}{n} \begin{bmatrix}
1 & 1 & 1 & \cdots & 1 \\
1 & \omega & \omega^2 & \cdots & \omega^{n-1} \\
1 & \omega^2 & \omega^4 & \cdots & \omega^{2(n-1)} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & \omega^{n-1} & \omega^{2(n-1)} & \cdots & \omega^{(n-1)(n-1)}
\end{bmatrix} $$
